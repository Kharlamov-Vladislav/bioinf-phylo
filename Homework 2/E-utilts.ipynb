```python
from Bio import Entrez
import warnings
import pandas as pd
from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')
```

```python
def esearch(name_organism, name_gene):
    response = []
    Entrez.email = "mike.rayko@gmail.com"
    handle = Entrez.esearch(db="nucleotide", term=f'"{name_organism}"[Organism] AND {name_gene}[Gene]')
    record = Entrez.read(handle)
    table_uid, table_acession, table_seqlenght = [], [], []
    for id in record['IdList']:
        handle = Entrez.efetch(db="nucleotide", id=id, rettype="gb", retmode="xml")
        response.append(handle.read())
        
        handle = Entrez.esummary(db="nucleotide", id=str(id))
        record = Entrez.read(handle)[0]
        table_uid.append(record['Id'])
        table_acession.append(record['Caption'])
        table_seqlenght.append(record['Length'])
    table_summary = pd.DataFrame({'UID' : table_uid, 'Acession' : table_acession, 'Sequance length' : table_seqlenght})
    return table_summary
def efetch(table_id):
    with open('efetch.fasta', 'a') as fasta:
        for id in table_id:
            Entrez.email = "mike.rayko@gmail.com"
            handle = Entrez.efetch(db="nucleotide", id="EU490707", rettype="fasta", retmode="text").read()
            fasta.write(handle)
def elink(pubmed_id):
    with open('elink.fasta', 'a') as fasta:
        Entrez.email = "mike.rayko@gmail.com"
        record = Entrez.read(Entrez.elink(dbfrom="pubmed", id=pubmed_id))
        for rec in record[0]['LinkSetDb'][0]['Link']:
            value = rec['Id']
            print(value)
            try:
                handle = Entrez.efetch(db="nucleotide", id=value, rettype="fasta", retmode="text").read()
                fasta.write(handle)
            except:
                continue # Some file too large for dowland
```
